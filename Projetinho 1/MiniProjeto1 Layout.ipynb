### Itens 4 a 6 - Resolução:

### Estimação dos parâmetros da curva normal

```python
mu = 120 # parâmetros média e desvio padrão estimados
sigma = 22
x=sorted(stats.norm.rvs(size=10000, loc=mu, scale=sigma))
```

```python
faixas = np.arange(80,300,5) #organiza as faixas por 5 minutos de filme
movies.duration.plot(kind='hist', bins = faixas, normed=True,color='black')
plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), lw=3, alpha=0.6, color='magenta')
plt.title('histograma da duração dos filmes em faixas de 5 minutos')
plt.xlabel('faixas')
plt.ylabel('densidade')
```

### Estimação dos parâmetros da curva de Weibull

```python
# k para a gamma é escolhido entre 1 e 2 baseado na forma do histograma e no efeito que k tem na forma da gamma
k_gamma = 30

# Estimamos theta da distribuição gamma
theta_gamma = movies.duration.mean()/k_gamma


x = faixas

#fig = plt.figure(figsize=(14,10))


# histograma e densidade teórica
movies.duration.plot(kind='hist', bins = faixas, normed=True,color='black')
gamma = stats.gamma.pdf(x, scale=theta_gamma, loc=0, a=k_gamma)


plt.plot(x, gamma, '-', color='green', label="Gamma") # Plota a gamma teórica

plt.legend()
```

### Fit das curvas acima

### Curva Normal

```python
Normal = stats.norm.fit(movies.duration) 
print(Normal)

mean, var, skew, kurt = stats.norm.stats(moments='mvsk')

x = faixas
plt.plot(x, stats.norm.pdf(x,  Normal[0],  Normal[1]), label='normal pdf')
plt.hist((movies.duration), 10, normed=1, facecolor='green', alpha=0.5)
plt.legend()
plt.show()
```

### Curva Gama

```python
Gamma = stats.gamma.fit(movies.duration) 
print(Gamma)

mean, var, skew, kurt = stats.norm.stats(moments='mvsk')

x = faixas
plt.plot(x, stats.gamma.pdf(x,  Gamma[0],  Gamma[1], Gamma[2]), label='gamma pdf')
plt.hist((movies.duration), 10, normed=1, facecolor='grey', alpha=0.5)
plt.legend()
plt.show()
```

### Itens 7 a 9 - Resolução:

```python
Normal = stats.norm.fit(movies.duration) 

mean, var, skew, kurt = stats.norm.stats(moments='mvsk')

x = faixas
plt.plot(x, stats.norm.pdf(x,  Normal[0],  Normal[1]), label='normal pdf')

Gamma = stats.gamma.fit(movies.duration) 

mean, var, skew, kurt = stats.norm.stats(moments='mvsk')

x = faixas
plt.plot(x, stats.gamma.pdf(x,  Gamma[0],  Gamma[1], Gamma[2]), label='gamma pdf')
plt.hist((movies.duration), 10, normed=1, facecolor='magenta', alpha=0.5)
plt.legend()
plt.show()
```

A curva gamma parece mais indicada do que a normal para descrever esse sistema, uma vez que sua forma se

### Q-Q Plot

```python
# Cálculo da frequencia relativa acumulada
acumulada = [] 
for i in range(len(movies.duration)):
    acc = (i - 0.5) / len(movies.duration)
    if acc < 0:
        acc = 0
    acumulada.append(acc) 


acumulada = pd.Series(data=sorted(acumulada))


quantil_normal = stats.norm.ppf(acumulada, loc=mu, scale=sigma) 
quantil_gamma = stats.gamma.ppf(acumulada, scale=theta_gamma, loc=0, a=k_gamma)


fig = plt.figure(figsize=(12,8))

plt.title("Q-Q plot: Dados x Normal x Gamma")

normal_plot, = plt.plot(sorted(movies.duration), quantil_normal, 'o', color="red", label="Normal")
gamma_plot, = plt.plot(sorted(movies.duration), quantil_gamma, 'o', color="green", label = "Gamma")

plt.legend(handles=[normal_plot, gamma_plot])
plt.plot((min(movies.duration), max(movies.duration)), (min(movies.duration), max(movies.duration)), '-')


for i in range(len(movies.duration)):
    plt.plot(movies.duration[i], movies.duration[i], (quantil_normal[i], movies.duration[i]), '-', color='red', alpha=0.3)  
    plt.plot(movies.duration[i], movies.duration[i], (quantil_gamma[i], movies.duration[i]), '-', color='green', alpha=0.3  )
```

### Cumulativas

```python
# fra e FDA
final = list(movies.duration)
final_boa = pd.Series(data=sorted(final))
cdf_normal = stats.norm.cdf(final_boa, loc=mu, scale=sigma) #FDA
cdf_gamma = stats.gamma.cdf(final_boa, scale=theta_gamma, loc=0, a=k_gamma)


fig = plt.figure(figsize=(12,8))

plt.title("Cumulativa: Dados x Normal x Gamma")

dados_fra_plot, = plt.plot(final_boa, acumulada, '-', color='black', label="F.R.A dos dados")
cdf_normal_plot,   = plt.plot(final_boa, cdf_normal, '-', color='red', label = "Cumulativa da exponencial")
cdf_gamma_plot, = plt.plot(final_boa, cdf_gamma, '-', color='green', label = "Cumulativa da distribuição Gamma")

plt.legend(handles=[dados_fra_plot, cdf_normal_plot, cdf_gamma_plot])


for i in range(len(final_boa)):
    plt.plot((final_boa[i], final_boa[i]), (acumulada[i], cdf_normal[i]), ':', color="red", alpha=0.3)
    #Omitido para não poluir muito o gráfico
    plt.plot((final_boa[i], final_boa[i]), (acumulada[i], cdf_gamma[i]), '--', color="green", alpha=0.3)
```

### Testes de Aderência

### Teste do Erro Quadrático:

```python
quantil_normal[0] = 120 #slicing nao deu certo, uma vez que não era possivel fazer operações com o último índice da lista
                        #final_boa
```

```python
erro_quadratico_normal = sum((final_boa - quantil_normal)**2)
erro_quadratico_gamma = sum((final_boa - quantil_gamma)**2)

print("Erro quadrático da normal :{:4.3f}\n".format(erro_quadratico_normal))
print("Erro quadrático da gamma :{:4.3f}\n".format(erro_quadratico_gamma))
```

### Teste do Erro Quadrático Médio

```python
eqm_normal = erro_quadratico_normal/len(final_boa)
eqm_gamma = erro_quadratico_gamma/len(final_boa)

print("Erro quadratico médio da normal :", eqm_normal)
print("Erro quadratico médio da gamma :", eqm_gamma)
```

### Soma de Diferenças

```python
soma_diferenças_normal = sum(abs(final_boa - quantil_normal))
soma_diferenças_gamma = sum(abs(final_boa - quantil_gamma))

print("Soma de diferenças da normal :", soma_diferenças_normal)
print("Soma de diferenças da gamma :", soma_diferenças_gamma)
```

### Teste de Kolmogorov-Smirnov

```python
# Kolmogorov-Smirnov goodness-of-fit test
ks_aderencia_normal = max(abs(acumulada-cdf_normal))
ks_aderencia_gamma = max(abs(acumulada-cdf_gamma))


print("Teste K-S para a normal com μ= {:4.2f} sigma={:4.3f} ks= {:4.3f}\n\
Teste K-S para a gamma com scale={:4.3f}, loc=0 e a={:4.3f} ks= {:4.3f}".format(Normal[0], Normal[1], ks_aderencia_normal, theta_gamma, k_gamma, ks_aderencia_gamma))
```
